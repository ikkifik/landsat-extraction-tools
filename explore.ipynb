```python
gcpdata = pd.read_csv('result_gcp.csv')
topbucket = gcpdata.sort_values(by=['date_acquired']).head()
```

```python
topbucket.reset_index()
topbucket.rename(columns={'Unnamed: 0': 'index'}, inplace=True)
topbucket.index = pd.RangeIndex(start=1, stop=len(topbucket.index)+1)

topbucket
```

```python
topbucket = topbucket.to_dict('records')

topbucket = pd.json_normalize(topbucket)
topbucket.to_json("result.json")
```

### Search Location of Indonesia

```python
gcpdf = pd.read_csv('result_gcp.csv')

search_path = gcpdf['wrs_path'].between(100, 140)
idn_bucket = gcpdf[search_path]
search_row = gcpdf['wrs_row'].between(55, 67)
idn_bucket = idn_bucket[search_row]
```

```python
gcpdf
```

```python
idn_bucket.reset_index()
idn_bucket.rename(columns={'Unnamed: 0': 'index'}, inplace=True)
idn_bucket.index = pd.RangeIndex(start=1, stop=len(idn_bucket.index)+1)

idn_bucket
```

```python
idn_bucket_dict = idn_bucket.to_dict('records')
```
